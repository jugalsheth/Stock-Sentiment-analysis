```python
import pandas as pd
```

```python
data = pd.read_csv('Data.csv',encoding = "ISO-8859-1")
```

```python
data.head(2)
```

```python
train = data[data['Date'] < '20150101']
test = data[data['Date'] > '20141231']
```

```python
traindf = train.iloc[:,2:27]
traindf.replace('[^a-zA-Z]',' ', regex = True, inplace = True)

list1 = [i for i in range(25)]
new_list = [str(i)for i in list1]
traindf.columns = new_list
```

```python
traindf.head(2)
```

```python
for i in new_list:
    traindf[i] = traindf[i].str.lower()
traindf.head(2)
```

```python
headline = []
for row in range(0,len(traindf.index)):
    headline.append(' '.join(str(x) for x in traindf.iloc[row,0:25]))
```

```python
#headline
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(ngram_range=(2,2))
trainingdataset = cv.fit_transform(headline)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=200,criterion='entropy')
rfc.fit(trainingdataset,train['Label'])
```

```python
test_transformer = []
for row in range(0,len(test.index)):
    test_transformer.append(' '.join(str(x) for x in test.iloc[row,2:27]))

test_data = cv.transform(test_transformer)
pred = rfc.predict(test_data)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(test['Label'],pred)
accuracy
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(test['Label'],pred)
cm
```

```python
from sklearn.metrics import classification_report
report = classification_report(test['Label'],pred)
report
```
